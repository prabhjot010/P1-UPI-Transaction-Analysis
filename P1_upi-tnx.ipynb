```python
## Q1: Transaction Overview
import pandas as pd
df = pd.read_csv(r"C:\Users\ASUS\OneDrive\DA\DATA SETS\upi_transactions_2024.csv")
df
```

```python
## Q2: Merchant Category Analysis
df['amount (INR)'].sum()
df['merchant_category'].value_counts().plot(kind='bar', color='brown')
plt.title('Merchant Category Transaction Count')
plt.xlabel('Category')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

```

```python
## Q3: Month-wise Transaction Trend
df.groupby('Month')['amount (INR)'].sum().plot(kind='line', color='brown')
plt.title('Monthly Transaction Count')
plt.xlabel('Month')
plt.ylabel('Amont')
plt.tight_layout()
plt.show()
```

```python
## Q4: Weekday vs Weekend Transactions
df['is_weekend'].value_counts().plot(kind='bar', color='coral')
plt.title('DAYS Transaction Count')
plt.xlabel('Days')
plt.ylabel('Amont')
plt.tight_layout()
plt.show()
```

```python
## Q5: Top Spending States
df.groupby('sender_state')['amount (INR)'].sum().plot(kind='line', color='brown')
plt.title('sender states ')
plt.xlabel('states')
plt.ylabel('Amont')
plt.tight_layout()
plt.show()
```

```python
## Q6: Most Used Sender Bank
df['sender_bank'].value_counts().plot(kind='line', color='brown')
plt.title('sender bank ')
plt.xlabel('bank')
plt.ylabel('Amont')
plt.tight_layout()
plt.show()
```

```python
## Q7: Fraud Analysis
df[df['fraud_flag'] == 1]['merchant_category'].value_counts().plot(kind='line', color='brown')
plt.title('Fraudd Analysis')
plt.xlabel('Category')
plt.ylabel('Amont')
plt.tight_layout()
plt.show()
```

```python
## Q8: Peak Transaction Hours
import matplotlib.pyplot as plt
df['hour_of_day'].value_counts().plot(kind='bar', color='brown')
plt.title('Peak TNX hours  ')
plt.xlabel('Amount')
plt.ylabel('Time')
plt.tight_layout()
plt.show()
```
